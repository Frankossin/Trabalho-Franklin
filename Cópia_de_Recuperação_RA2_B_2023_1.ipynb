# Questão 1
Derivem as funções a seguir. 
Para todas as derivadas apresente a álgebra necessária explicitando matematicamente, com latex, a aplicação de todas as regras, simplificações necessárias. Não deixe de plotar os gráficos das funçoes e das derivadas usando o matplotlib.

## 1. 
$$f(x) = x^3 + 5x^2 - 7x + 9$$

## 3
$$g(x) = e^x \cos(x)$$


```python
#Função 1
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função original e sua derivada
def f(x):
    return x**3 + 5*x**2 - 7*x + 9

def df(x):
    return 3*x**2 + 10*x - 7

# Gerando valores para x
x = np.linspace(-10, 10, 400)

# Calculando y para a função original e sua derivada
y = f(x)
dy = df(x)

# Criando o gráfico
plt.figure(figsize=(12, 6))

# Plotando a função original
plt.subplot(1, 2, 1)
plt.plot(x, y, label='$f(x) = x^3 + 5x^2 - 7x + 9$')
plt.xlabel('x')
plt.ylabel('$f(x)$')
plt.grid(True)
plt.legend()

# Plotando a derivada
plt.subplot(1, 2, 2)
plt.plot(x, dy, label="$f'(x) = 3x^2 + 10x - 7$", color='black')
plt.xlabel('x')
plt.ylabel("$f'(x)$")
plt.grid(True)
plt.legend()

# Mostrando o gráfico
plt.tight_layout()
plt.show()
```

Passo a passo da 1

1. A derivada de uma constante é zero, então a derivada de 9 é 0.

   \begin{equation}
   \frac{d}{dx}9 = 0
   \end{equation}

2. A derivada de \(x\) com respeito a \(x\) é 1, então a derivada de \(-7x\) é \(-7\).

   \begin{equation}
   \frac{d}{dx}(-7x) = -7
   \end{equation}

3. A derivada de \(x^2\) é \(2x\), então a derivada de \(5x^2\) é \(5 * 2x = 10x\).

   \begin{equation}
   \frac{d}{dx}(5x^2) = 10x
   \end{equation}

4. A derivada de \(x^3\) é \(3x^2\).

   \begin{equation}
   \frac{d}{dx}(x^3) = 3x^2
   \end{equation}

Somando todas essas derivadas, obtemos a derivada da função original:

\begin{equation}
\frac{d}{dx}(x^3 + 5x^2 - 7x + 9) = 3x^2 + 10x - 7
\end{equation}

```python
#Função 2
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função e sua derivada
def g(x):
    return np.exp(x) * np.cos(x)

def dg(x):
    return np.exp(x) * np.cos(x) - np.exp(x) * np.sin(x)

# Gerando valores para x
x = np.linspace(-2*np.pi, 2*np.pi, 1000)

# Calculando os valores de y para a função e sua derivada
y = g(x)
dy = dg(x)

# Plotando a função e sua derivada
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='$g(x) = e^x cos(x)$')
plt.plot(x, dy, label="$g'(x) = e^x cos(x) - e^x sin(x)$")
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Função e sua derivada')
plt.grid(True)
plt.show()
```

Passo a passo da 2

$$g(x) = e^x \cos(x)$$

Utilizando a regra do produto:
$$g(x) = e^x * \cos(x)+ e^x * \frac{d}{dx}(\cos(x))$$

Aplicando a regra da potência e da diferença:
$$g(x) = e^x * \cos(x)+ e^x * -(\sin(x))$$

Simplificando:
$$g(x) = e^x( \cos(x) - \sin(x))$$

$$j(x) = \sqrt{x} \ln(x)$$

$$j(x) = [\sqrt{x}]* \ln(x)+  \sqrt{x}* \frac{d}{dx}\ln(x)$$
Aplicando a regra da potência e diferença:
$$\frac{1}{2}x^{\frac{1}{2}-1}* \ln(x)+  \sqrt{x}*\frac{1}{x} $$

# Questão 3
Usando o python, numpy, encontre o máximo local de $$f(x)=x^2-98x+4$$ para $0 \leq x \leq 100$. 
 

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 - 98*x + 4

# Encontrar o máximo local
x_max_local = None
f_max_local = float('-inf')

for x in range(0, 101):
    f_val = f(x)
    if f_val > f_max_local:
        f_max_local = f_val
        x_max_local = x

print("O máximo local da função ocorre em x =", x_max_local)
print("O valor máximo local é f(x) =", f_max_local)

x_vals = np.linspace(0, 100, 1000)
y_vals = f(x_vals)

y_max_local = f(x_max_local)

plt.plot(x_vals, y_vals, label='$f(x) = x^2 - 98x + 4$')
plt.scatter(x_max_local, y_max_local, color='red', label='Máximo Local')

plt.title('$f(x) = x^2 - 98x + 4$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

# Questão 1 (vale até 2 pontos)
Determine a área entre as curvas $$y=xe^{-x^{2}}$$ $$y=x+1$$ $$x=2$$ e o eixo $y$, usando somente matplotlib, numpy, sympy. Trace o gráfico desta área preenchendo a área entre as curvas em uma cor diferente.



```python

import numpy as np
import matplotlib.pyplot as plt
from sympy import *

x = symbols('x')
f_symp = x * exp(-x**2)
g_symp = x + 1
integral_f = integrate(f_symp,x)
integral_g = integrate(g_symp,x)
area = integral_f - integral_g

f = lambdify(x,f_symp,'numpy')
g = lambdify(x,g_symp,'numpy')

x_val = np.linspace(0,2,1000)
f_val = f(x_val)
g_val = g(x_val)

plt.figure(figsize=(10,6))
plt.plot(x_val,f_val)
plt.plot(x_val,g_val)

plt.fill_between(x_val, f_val, g_val, where=(f_val<g_val), color='skyblue', alpha=0.3)
plt.grid(True)
plt.show()
```

# Questão 2 (vale até 2 pontos)
Resolva, algebricamente, explicitando em latex todos os passos necessários, as seguintes integrais:

##a) 
$$\int \cos(3t) \sin(8t) dt $$
##b)
$$\int 4x \cos(2-3x) dx$$


##a) 
$$\int \cos(3t) \sin(8t) dt $$
\begin{align*}
\text{Usando a identidade trigonométrica:} \quad \sin(a)\cos(b) &= \frac{1}{2}[\sin(a+b) + \sin(a-b)] \\
\int \cos(3t)\sin(8t) \, dt &= \frac{1}{2} \int [\sin(3t+8t) + \sin(3t-8t)] \, dt \\
&= \frac{1}{2} \int [\sin(11t) + \sin(-5t)] \, dt \\
&= \frac{1}{2} \left[-\frac{1}{11} \cos(11t) - \frac{1}{5} \cos(5t)\right] + C \\
&= -\frac{ \cos(11t)}{22} - \frac{\cos(5t)}{10} + C
\end{align*}


##b)
$$\int 4x \cos(2-3x) dx$$
\begin{align*}
\text{Fazendo a regra da substituição:} \quad u &= 2-3x, \quad du = -3 \, dx \\
\int 4x\cos(2-3x) \, dx &= \int -\frac{4}{3}\cos(u) \, du \\
&= -\frac{4}{3} \int \cos(u) \, du \\
&= -\frac{4}{3} \sin(u) + C \\
&= -\frac{4}{3} \sin(2-3x) + C
\end{align*}


# Questão 3 (vale até 4 pontos)
Calcule o valor das integrais definidas explicitadas a seguir. Trace o gráfico das funções que foram integradas e das funções encontradas após a integração destacando, com linhas verticais, os limites de integração. 

##a)
$$\int_0^2 x^2+1 \space dx$$

##b) 
$$\int_{-3}^1 6x^2-5x+2 \space dx$$



```python
#a)
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, sin, integrate, lambdify

x = symbols('x')
f_symp = x**2 + 1

integral = integrate(f_symp,(x,0,2))
integralform = integrate(f_symp,x)
print(integralform)
integral_val = float(integral)

print(f"valor da integral definida: {integral_val}")

f = lambdify(x,f_symp,'numpy')
f_integral = lambdify(x,integralform,'numpy')

x_val = np.linspace(0,10,1000)
f_val = f(x_val)
f_integral_val = f_integral(x_val)

plt.figure(figsize=(10,6))
plt.plot(x_val,f_val, label = "$f(x) = x^2 + 1$")
plt.plot(x_val, f_integral_val, label = '$x^3/3 + 3$')
plt.axvline(x=0, color='blue', linestyle='--')
plt.axvline(x=2, color='blue', linestyle='--')

idx = np.where((x_val >= 0) & (x_val <= 2))
plt.fill_between(x_val[idx], f_val[idx], f_integral_val[idx], color='skyblue', alpha=0.3)
plt.legend()
plt.grid(True)
plt.show()
```

```python
#b)
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, sin, integrate, lambdify

x = symbols('x')
f_symp = 6*x**2 -5*x + 2

integral = integrate(f_symp,(x,-3,1))
integralform = integrate(f_symp,x)
print(integralform)
integral_val = float(integral)

print(f"valor da integral definida: {integral_val}")

f = lambdify(x,f_symp,'numpy')
f_integral = lambdify(x,integralform,'numpy')

x_val = np.linspace(-10,10,1000)
f_val = f(x_val)
f_integral_val = f_integral(x_val)

plt.plot(x_val,f_val, label = "$f(x) = 6x^2 - 5x + 2$")
plt.plot(x_val, f_integral_val, label = '$ 2x^3 - (5x^2)/2 + 2x$')
plt.axvline(x=-3, color='blue', linestyle='--')
plt.axvline(x=1, color='blue', linestyle='--')

idx = np.where((x_val >= -3) & (x_val <= 1))
plt.fill_between(x_val[idx], f_val[idx], f_integral_val[idx], color='skyblue', alpha=0.3)
plt.legend()
plt.grid(True)
plt.show()
```

#Questão 4 (vale até 1 ponto)
A taxa de variação de juros nos primeiros meses do ano de 2023 é dada por $v = 3t^2 - e^t$. Onde $t$ é o tempo dado em dias. Encontre a função que determina o juro, $j$ a ser pago em um determinado dia deste período de forma algébrica em latex.



Resposta:
\begin{align*}
\text{Dada a taxa de variação de juros:} \quad v(t) &= 3t^2 - e^t \\
\text{ integramos em relação ao tempo:} \quad j(t) &= \int v(t) \, dt \\
&= \int (3t^2 - e^t) \, dt \\
&= \int 3t^2 \, dt - \int e^t \, dt \\
&= t^3 - \int e^t \, dt \\
&= t^3 - e^t + C \\
\end{align*}

#Questão 5 (vale até 1 ponto)
Encontre o valor da constante C para a antiderivada $F(x)$ da função $f(x) = 3x^3 - 2x + 1$ de tal forma que $F(1)=0$. Apresente seus cálculos em latex e explique as decisões que tomou. 

Resposta:
\begin{align*}
\text{Dada a função } f(x) &= 3x^3 - 2x + 1 \\
\text{Antiderivada } F(x) &= \int f(x) \, dx \\
&= \int (3x^3 - 2x + 1) \, dx \\
&= \frac{3}{4}x^4 - x^2 + x + C \\
\\
F(1) &= \frac{3}{4}(1)^4 - (1)^2 + (1) + C \\
0 &= \frac{3}{4} - 1 + 1 + C \\
0 &= \frac{3}{4} + C \\
C &= -\frac{3}{4}
\end{align*}

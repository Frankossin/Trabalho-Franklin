 Considerando as integrais a seguir você deverá plotar o gráfico de cada uma das equações que estão
 sendo integradas e marcar a área entre os limites de integração. Depois, usando apenas álgebra,
 calcular a área sobre estas curvas e, finalmente, verificar seus cálculos calculando a área sobre estas
 curvas usando a Soma de Reimann. Esta Tarefa deve ser realizada no Google Colab usando o Python,
limitado ao uso das bibliotecas numpy, sympy e matplotlib

```python

#Alunos Franklin J.Belasque Bauch Vieira, Eduardo de Abreu, Caike Augusto, Vinicius Zaia
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, sin, integrate, lambdify

x = symbols('x')
f_symp = 40 * x**3 + 12 * x**2 - 9 * x 

integral = integrate(f_symp,(x,2.5,7))
integral_val = float(integral)

print(f"valor da integral definida: {integral_val}")


#Soma de reimann para verificar

# Definindo a função
def funcao(x):
    return 40 * x**3 + 12 * x**2 - 9 * x 

# Parâmetros para a soma de Riemann
limite_inferior = 2.5  
limite_superior = 7  
numero_retangulos = 100  

# Calculando a largura de cada retângulo
largura_retangulo = (limite_superior - limite_inferior) / numero_retangulos

# Calculando a soma de Riemann
soma_riemann = 0
for i in range(numero_retangulos):
    # O x no topo do retângulo
    x_topo_retangulo = limite_inferior + i * largura_retangulo
    # Altura do retângulo
    altura_retangulo = funcao(x_topo_retangulo)
    # Área do retângulo
    area_retangulo = largura_retangulo * altura_retangulo
    # Adicionando a área ao total
    soma_riemann += area_retangulo

print("A soma de Riemann é", soma_riemann)


f = lambdify(x,f_symp,'numpy')

x_val = np.linspace(0,10,1000)
f_val = f(x_val)

plt.figure(figsize=(10,6))
plt.plot(x_val,f_val,label = '$I(t) = 20 * \sin (t) + 50 $')

plt.axvline(x=2.5, color='blue', linestyle='--')
plt.axvline(x=7, color='blue', linestyle='--')

idx = np.where((x_val >= 2.5) & (x_val <= 7))
plt.fill_between(x_val[idx], f_val[idx], color='skyblue', alpha=0.3)
plt.grid(True)
plt.show()


```

Calculo da área

$$\int_{1}^{2} \frac{2x^5 - x + 3}{x^2} \,\, dx$$

$$ Integral$$
$$ -\ln(x) + \frac{x^4}{2} - \frac{3}{x} $$

$$ Trocando \,\, por\,\, 1$$
$$ -\ln(1) + \frac{1^4}{2} - \frac{3}{1} $$
$$ = -2.5$$

$$ Trocando \,\, por\,\, 2$$
$$-\ln(2) + \frac{2^4}{2} - \frac{3}{2} $$
$$ = 5.80$$

$$ 5.80 -(-2.5) = 8.3$$

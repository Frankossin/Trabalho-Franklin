```python
import numpy as np
import sympy as sp
import math
import matplotlib.pyplot as plt
from IPython.display import display, Math
from fractions import Fraction
```

```python
#Ex 1
#set symbols and equations
x1 =sp.Symbol("x")
y2 = (x1**2 - 4)/(x1 - 2)
x = np.linspace(-10,10,10)
y = (x**2 - 4)/(x - 2)
lim_3 = sp.limit(y2,x1,3)
lim_0 = sp.limit(y2,x1,0)

#passo a passo
display(Math(r"\lim{x \to 3} f(x) = " + f"{y2}"))
display(Math(r"\lim{x \to 3} f(x) = " + f"{y2.subs(x1,3)}"))

#plot
plt.plot(x,y)
plt.plot(3,5,"o",color="red",label = "Limite esquerdo e direito")
plt.title(r'$\mathrm{Gr√°fico\ de}\ f(x) = \frac{x^2 - 4}{x - 2}$')
plt.legend()
plt.text(2,5.5,"$(3,5)$")
```

```python
#c)
print("c)")

#symbols and equations
x = sp.Symbol('x')
f = ((2*x)**2 - 5*x + 3)/((2*x)**2 - 3*x)
ff = sp.factor(f)
f = (4*x - 5 + 3/x) / (4*x - 3)
f_subs = f.subs(x, sp.Rational(3, 2))
limite = sp.limit(f_subs, x, sp.Rational(3, 2))
limite_fracao = Fraction(limite).limit_denominator()

#passo a passo
display(Math(r"$\lim_{x \to 3/2} f(x) = \frac{2x^2 - 5x + 3}{2x^2 - 3x}$"))
print("\n")
display(ff)
print("\n")
display(Math(r"$\lim_{x \to 3/2} f(x) = \frac{x(4x - 5 + 3/x)}{x(4x - 3)}$"))
print("\n")
display(Math(r"$\lim_{x \to 3/2} f(x) = \frac{(4x - 5 + 3/x))}{4x - 3}$"))
print("\n")
display(Math(r"\lim_{x \to 3/2} f(x) = "+f"{limite_fracao}"))
print("\n")

#plot
x3 = np.linspace(8,20,200)
y_vals2 = np.array([sp.N(f.subs(x, val)) for val in x3])
plt.plot(x3,y_vals2)
plt.show()
```

# Questão 1
Uma doença se espalha de forma exponencial, com uma taxa de contágio de 0,05 por dia. Se inicialmente 1000 pessoas são infectadas, faça um gráfico que mostre como o número de pessoas infectadas varia com o tempo e determine quantas pessoas estarão infectadas após 30 dias. Utilize a equação $I = I_0 * e^{rt}$, onde $I_0$ é o número inicial de infectados, `r` é a taxa de contágio diária (em decimal) e $t$ é o tempo decorrido em dias.



```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Math

t = np.linspace(0,30,200)
I = 1000 * np.e**(0.05*t)

res =int( 1000*np.e**(0.05*30))

plt.plot(t,I)
print("Resultado após 30 dias = "+f"{res}")
```

# Questão 2
Usando, a seu critério, as bibliotecas do Python, crie um gráfico para cada uma das funções a seguir. No label do gráfico deverá estar a função impressa com qualidade matemática (latex) e uma label dizendo se a função é crescente ou decrescente. 

1. $f(x) = 3x^2 - 6^x +48$

2. $f(x) = \log_{10}(x^2 - 3x) - \log_{10}(x - 3)$

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
from math import log
from IPython.display import display, Math

#funcao 1
print("Função 1")
x = np.linspace(-12,12,100)
y = (3*(x**2) - (6**x) + 48)
plt.plot (x,y)
plt.text(-8, -100, "Decrescente")
plt.title(r"$f(x) = 3x^2-6^x+48$")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
from math import log
from IPython.display import display, Math

#funcao 2
print("Função 2")
x = sy.Symbol("x")
y2 = sy.log((x**2)-(3*x), 10)- sy.log(x-3, 10)
x_vals = np.linspace(3.01, 12, 100)  
y_vals = np.array([sy.N(y2.subs(x, val)) for val in x_vals])  
plt.plot(x_vals, y_vals)
plt.title(r"$f(x) = \log_{10}(x^2-3x)-\log_{10}(x-3)$")
plt.text(6,1, r"$Crescente$")
```

1. $\lim_{x \to 1} \frac{x^{10} - 1}{x^5 - 1}$

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
from math import log
from IPython.display import display, Math

#1
print("1.")

display(Math(r"$\lim{x \to 1} f(x) = \frac{x^{10}-1}{x^5 - 1}$"))
print("\n")
display(Math(r"$\lim{x \to 1} f(x) = \frac{(x^{5}-1)(x^5 + 1)}{x^5 - 1}$"))
print("\n")
display(Math(r"$\lim{x \to 1} f(x) = x^{5}+1$"))
print("\n")
display(Math(r"$\lim{x \to 1} f(x) = 2$"))
print("\n")

x1 = np.linspace(-0.07,2.1,100)
y1 = ((x1**10) - 1)/((x1**5) -1)
plt.plot(1,2,"o",color="black")
plt.plot(x1,y1)
plt.show()
print("\n")

#2
print("2.")
print("Limite da esquerda")
display(Math(r"$\lim{x \to \pi-} f(x) = \frac{cos(x)-1}{x - \pi}$"))
print("\n")
display(Math(r"$\lim{x \to \pi-} f(x) =(cos(x)-1) \frac{1}{x - \pi}$"))
print("\n")
display(Math(r"$\lim{x \to \pi-} f(x) =\lim{x \to \pi-}(cos(x)-1)* \lim{x \to \pi-}\frac{1}{x - \pi}$"))
print("\n")
display(Math(r"$\lim{x \to \pi-} f(x) = -2 * -\infty$"))
print("\n")
display(Math(r"$\lim{x \to \pi-} f(x) =\infty$"))

print("\n\nLimite da direita\n")
display(Math(r"$\lim{x \to \pi+} f(x) = \frac{cos(x)-1}{x - \pi}$"))
print("\n")
display(Math(r"$\lim{x \to \pi+} f(x) =(cos(x)-1) \frac{1}{x - \pi}$"))
print("\n")
display(Math(r"$\lim{x \to \pi+} f(x) =\lim{x \to \pi+}(cos(x)-1)* \lim{x \to \pi+}\frac{1}{x - \pi}$"))
print("\n")
display(Math(r"$\lim{x \to \pi+} f(x) = -2 * \infty$"))
print("\n")
display(Math(r"$\lim{x \to \pi+} f(x) =-\infty$"))
print("\n")

def f(x):
    return (np.cos(x) - 1)/(x - np.pi)

x2 = np.linspace(np.pi - 0.5, np.pi + 0.5, 100)
y2 = f(x2)

plt.plot(x2, y2)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r'Gráfico de $f(x) = \frac{\cos(x) - 1}{x - \pi}$')
plt.show()
```
